## Integración Montecarlo

1. Realizar las siguientes integrales. Poner los resultados en una tabla cuyas filas correspondan a la cantidad de términos utilizados en la aproximación (usar 10, 100, 1000, 10000 y 100000 términos) y cuyas columnas correspondan a las funciones.
    + $\int_{1}^{3} e^{x}\log(x)\text{d}x$.
    + $\int_1^3 \sin(x)\ln(x) \text{d}x$

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
# Para la primera Función
b=3
a=1
def f1(x):
    return (np.exp(x)*np.log10(x))

def montecarlo1(f1,N,a,b):
    u=np.random.uniform(a,b,N)
    return np.mean(f1(u))*(b-a)
N=np.logspace(1,5,5)
tabla1=pd.DataFrame(columns=["Aprox. Función 1","Error Función 1"], index=N)

for i in N:
    aprox=montecarlo1(f1,int(i),a,b)
    tabla1.loc[i,"Aprox. Función 1"]=aprox
    tabla1.loc[i,"Error Función 1"]=(14.0275)-aprox
```

```python
# Para la segunda Función

def f2(x):
    return (np.sin(x)*np.log(x))

def montecarlo2(f2,N,a,b):
    u=np.random.uniform(a,b,N)
    return np.mean(f2(u))*(b-a)
N=np.logspace(1,5,5)
tabla2=pd.DataFrame(columns=["Aprox. Función 2","Error Función 2"], index=N)

for i in N:
    aprox=montecarlo1(f2,int(i),a,b)
    tabla2.loc[i,"Aprox. Función 2"]=aprox
    tabla2.loc[i,"Error Función 2"]=(0.869844)-aprox
```

```python
pd.concat([tabla1,tabla2],axis=1)
```

2. Aproximar el área de la región descrita por

$$x^2+2y^2\leq-2x+8y-7$$

Poner los resultados en una tabla cuyas filas correspondan a la cantidad de términos utilizados en la aproximación (usar 10, 100, 1000, 10000 y 100000 términos) y en las columnas el valor de aproximación y el error relativo.

Definir unos límites adecuados $a_1$, $b_1$, $a_2$ y $b_2$, para la región $R=\left[a_1,b_1\right]\times\left[a_2,b_2\right]$

```python
def fun2(x,y):
    return (x**2)+2*(y**2) <= (-2*x)+8*y-7
```

```python
x=np.random.uniform(-2.6,.6,100)
y=np.random.uniform(.5,3.2,100)

X,Y=np.meshgrid(x,y)

plt.figure()
plt.scatter(X,Y,c=~fun2(X,Y))
```

```python
def montecarlito(region,ax,bx,ay,by,N):
    x=np.random.uniform(ax,bx,N)
    y=np.random.uniform(ay,by,N)
    
    B=region(x,y)
    AR=(bx-ax)*(by-ay)
    
    return np.mean(B)*AR
```

```python
tabla_act=pd.DataFrame(columns=["Aproximación"],index=N)
```

```python
for i in N:
    aprox=montecarlito(fun2,-2.5,0.5,0.5,3,int(i))
    tabla_act.loc[i,"Aproximación"]=aprox
tabla_act
```
