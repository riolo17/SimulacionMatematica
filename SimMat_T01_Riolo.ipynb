## La Nanotecnología en Recubrimientos

La nanotecnología encontró una gran oportunidad de aprovechamiento en el campo de los recubrimientos con aplicaciones diversas dentro de las industrias farmaceuticas, mecánicas, de alimentos, aeroespacial, óptica, tintas o pinturas, etc. Por ejemplo, en la industria farmaceuticas se usan nanopartículas de $TiO_{2}$ en bloqueadores para hacer capas protectoras de luz UV. Otro ejemplo muy común en distintas industrias, son los recubrimientos con nanopelículas de platas; este nanomaterial tiene propiedades antimicrobianas, siendo de gran utilidad en el campo de la salud y de el envase de alimentos.

Lo complicado al momento de crear estos materiales recae en las técnicas utilizadas. Por lo general, se pueden dividir en dos: Top-Down o Bottom-Up.
